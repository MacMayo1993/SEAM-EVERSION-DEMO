# Interactive SEAM VIZ Eversion Demo

Slide t to move through the manifold. Watch the surface warp, corrugate, and evert!

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interact, FloatSlider
import sys
sys.path.append('..')
from src.geometry import get_continuous_geometry

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim(-2,2); ax.set_ylim(-2,2); ax.set_zlim(-2,2)
ax.axis('off')

def update(t=0.0, k_star=1/(2*np.log(2))):
    ax.cla()
    ax.set_xlim(-2,2); ax.set_ylim(-2,2); ax.set_zlim(-2,2)
    ax.axis('off')
    
    X, Y, Z, C = get_continuous_geometry(1.0, t)  # Full sphere mode
    surf = ax.plot_surface(X, Y, Z, facecolors=C, shade=True)
    
    z_val = np.exp(-np.abs(t) / k_star)
    ax.set_title(f't = {t:.2f}    z(t) = {z_val:.3f}    bits resolved = {-np.log2(z_val):.1f}')
    plt.draw()

interact(update,
         t=FloatSlider(min=-5, max=5, step=0.1, value=0, description='t'),
         k_star=FloatSlider(min=0.3, max=1.5, step=0.05, value=1/(2*np.log(2)), description='k*'));
```
